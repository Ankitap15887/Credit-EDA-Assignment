#### defining plyanalysis for visual analysis

```python

import seaborn as sns

def plyanalysis(ds,col,title,hue = None):
        
    sns.set_style('white')
    sns.set_context('poster')
    plt.rcParams["axes.labelsize"] = 15
    plt.rcParams['axes.titlesize'] = 18
    plt.rcParams['axes.titlepad'] = 30
    
    plot = pd.Series(data = hue)
    fig, ax = plt.subplots()
    width = len(ds[col].unique()) + 7 + 4*len(plot.unique())
    fig.set_size_inches(width , 8)
    
    plt.title(title,fontsize=14)
    ax = sns.countplot(data=ds, x=col, hue=hue, palette='rainbow') 
    plt.setp(ax.get_xticklabels(), rotation=45, ha="right", rotation_mode="anchor")

    plt.show()
```

## bivariate analysis function


```python
def plbycat(cat, num):

    plt.style.use('ggplot')
    sns.despine
    fig,ax = plt.subplots(1,1,figsize=(10,5))
    
    sns.boxplot(x=cat,y = num, data=ds1)
    ax.set_ylabel(f'{num}')
    ax.set_xlabel(f'{cat}')

    ax.set_title(f'{cat} Vs {num}',fontsize=15)
    ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="left")
     
    plt.show()
```
